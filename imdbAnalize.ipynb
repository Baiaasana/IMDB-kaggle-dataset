## Simple app that recomend movies based on the content of the IMDB movie dataset by director/ganre/overview column values

```python
#checking null values
movies.isna().sum()
```

```python
# drop null values and working with selected features
movies.dropna(axis=0)
movies_data = movies[['Series_Title','Genre','IMDB_Rating','Overview','Poster_Link','Director']]
movies_data.head()
# movies_data['tags'] = movies_data['Overview'] +' '+ movies_data['Genre'] + ' ' + movies_data['Director']
```

```python
#adding the rows and make a new column
# movies_data['tags'] = movies_data['Overview'] +' '+ movies_data['Genre'] + ' ' + movies_data['Director']
#movies_data.loc[:, 'tags'] = movies_data['Overview'] +' '+ movies_data['Genre'] + ' ' + movies_data['Director']

# Creating the 'tags' column
movies_data['tags'] = movies_data['Overview'] + ' ' + movies_data['Genre'] + ' ' + movies_data['Director']

# Displaying the DataFrame
print(movies_data.head())
```

```python
movies_data.drop(['Overview'],axis=1,inplace=True)
```

```python
movies_data.head()
```

```python
# apply tfidf
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(stop_words='english',max_features = 10000)
vector = tf.fit_transform(movies_data['tags'])
```

```python
#findout the cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
distance = sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1]) 

for i in distance[0:5]:
    print(movies_data.iloc[i[0]].Series_Title , movies_data.iloc[i[0]].Genre)
```

```python
#make a function that recomend movie by movie title

def recommend(movie):
    index = movies_data[movies_data['Series_Title']==movie].index[0]
    distance = sorted(list(enumerate(similarity[index])),reverse=True,key=lambda vector:vector[1])
    for i in distance[1:10]:
        print(f"{movies_data.iloc[i[0]].Series_Title}\t\t\t{movies_data.iloc[i[0]].Genre}")
```

```python
#check 
recommend("Interstellar")
```
